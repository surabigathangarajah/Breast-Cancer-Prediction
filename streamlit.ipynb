```python

!pip install streamlit streamlit_jupyter joblib

import streamlit as st
from streamlit_jupyter import StreamlitPatcher, tqdm
import joblib
import numpy as np

# Enable Streamlit in notebook
StreamlitPatcher().jupyter()

model = joblib.load('/kaggle/input/breastcancer/breast_cancer_model.pkl')
print("Model loaded successfully ✅")
# Streamlit App
st.title("Breast Cancer Prediction App")
st.write("Enter tumor features to predict whether it is Malignant (Cancer) or Benign (No Cancer)")

# Input features (simplified: you can add all 30 features later)
radius_mean = st.number_input("Radius Mean", 0.0, 50.0, 14.0)
texture_mean = st.number_input("Texture Mean", 0.0, 50.0, 20.0)
perimeter_mean = st.number_input("Perimeter Mean", 0.0, 200.0, 90.0)

# Prediction button
if st.button("Predict"):
    features = np.array([[radius_mean, texture_mean, perimeter_mean]])
    prediction = model.predict(features)[0]
    if prediction == 1:
        st.error("Malignant (Cancer)")
    else:
        st.success("Benign (No Cancer)")





```
